### Operadores
```python
conjunto.add(x)
```

````python
# Prueba estos operadores e indica su función
# TODO


conjunto1 = {1, 2, 3}
conjunto2 = {3, 4, 5}

# conjunto.add(x): Agrega el elemento x al conjunto
conjunto1.add(6)
print("conjunto1 después de add(6):", conjunto1)

# conjunto.remove(x): Elimina el elemento x del conjunto. Lanza una excepción KeyError si x no está presente.
conjunto1.remove(2)
print("conjunto1 después de remove(2):", conjunto1)

# conjunto.discard(x): Elimina el elemento x del conjunto. No lanza una excepción si x no está presente.
conjunto1.discard(2)
print("conjunto1 después de discard(2):", conjunto1)

# conjunto.pop(): Elimina y devuelve un elemento arbitrario del conjunto.
elemento_extraido = conjunto1.pop()
print("Elemento extraído de conjunto1 usando pop:",elemento_extraido)
print("conjunto1 después de pop:", conjunto1)

# conjunto.update(conjunto): Agrega todos los elementos del conjunto dado al conjunto actual
conjunto1.update({2, 7})
print("conjunto1 después de update({2, 7}):", conjunto1)

# x in conjunto: Verifica si x está presente en el conjunto, devuelve True o False
print("¿2 está en conjunto1?:", 2 in conjunto1)

# conjunto1 <= conjunto2: Verifica si conjunto1 es subconjunto de conjunto2
print("¿conjunto1 es subconjunto de conjunto2?:", conjunto1 <= conjunto2)

# conjunto1 >= conjunto2: Verifica si conjunto1 es superconjunto de conjunto2
print("¿conjunto1 es superconjunto de conjunto2?:", conjunto1 >= conjunto2)

# conjunto1 - conjunto2: Diferencia de conjuntos, elementos en conjunto1 pero no en conjunto2
print("Diferencia entre conjunto1 y conjunto2:", conjunto1 - conjunto2)

# conjunto1 | conjunto2: Unión de conjuntos, elementos en conjunto1 o en conjunto2
print("Unión de conjunto1 y conjunto2:", conjunto1 | conjunto2)

# conjunto1 & conjunto2: Intersección de conjuntos, elementos comunes en conjunto1 y conjunto2
print("Intersección de conjunto1 y conjunto2:", conjunto1 & conjunto2)

# conjunto1 ^ conjunto2: Diferencia simétrica, elementos que están en conjunto1 o en conjunto2, pero no en ambos
print("Diferencia simétrica entre conjunto1 y conjunto2:", conjunto1 ^ conjunto2)
````

## Variables y métodos privados
No existen en python, no obstante hay una convención que dice que si una variable esta precedida de dos guiones bajos, es una variable privada.

```python
class NombreClase:
    def __init__(self, argumentos):

    def setVariablePrivada(self, argumento):

```

### Property: Analogo a getter y setter
```python
class NombreClase:
    def __init__(self, argumentos):
        self.__variablePrivada = argumento
    
    @property #decorador getter
    def variablePrivada(self):
        return self.__variablePrivada
        
    @variablePrivada.setter # decorador setter
    def variablePrivada(self, argumento):
        self.__variablePrivada = argumento
```

````python
# TODO Crear una clase cuyo metodo constructor reciba un diccionario
# como argumento e inicialice tantas variables del objecto como
# elementos tenga el diccionario. La clave sera el nombre de la
# variable y el valor su valor.
# self.valor_clave contiene valor del diccionario


class MiClase:
    def __init__(self, diccionario):
        for clave, valor in diccionario.items():
            setattr(self, clave, valor)

diccionario_ejemplo = {"variable1": "Manu", "variable2": 8, "variable3": True}
objeto = MiClase(diccionario_ejemplo)

print(objeto.variable1)
print(objeto.variable2)
print(objeto.variable3)
````
