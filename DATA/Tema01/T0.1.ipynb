```python
dic = {1:"uno", 2:"dos", 3:"tres"}
```

```python
dic2 = {1:["uno", "one"], 2:["dos", "two"], 3:["tes", "three"]}
```

```python
print(dic[2])
```

```python
while True:
    try:
        numero = int(input("Ingresa un número (1, 2, o 3): "))
        if numero in dic2:
            idioma = input("¿Quieres el valor en español o inglés? (es/en): ").lower()
            if idioma == 'es':
                resultado = dic2[numero][0]
                print(f"El valor correspondiente es: {resultado}")
                break
            elif idioma == 'en':
                resultado = dic2[numero][1]
                print(f"The corresponding value is: {resultado}")
                break
            else:
                print("Por favor, ingresa 'es' o 'en' para seleccionar el idioma.")
        else:
            print("Por favor, ingresa un número válido (1, 2, o 3).")
    except ValueError:
        print("Por favor, ingresa un número entero.")
```

```python
num = 0

while num not in dic2.keys():
  num = int(input("Dame un numero: "))

  leng = input("En que lenguaje? es/en: ")
  if leng == "es":
    pos = 0
  else:
    pos = 1
  print(dic2[num][pos])
```
