#### 1. Load Excel spreadsheet

Load the file `../datasets/pandas_homework1.xlsx` which contains a table with body measures from 200 people. Set the column `Name` as index.  
Display the table to understand its structure.

```python
import pandas as pd

file_path = "pandas_homework1.xlsx"
data = pd.read_excel(file_path, index_col="Name")

print(data)
```

#### 2. Table summary
Calculate the count, mean, standard deviation, minimum, maximum and quartiles of `Height` and `Weight`.

```python
height_stats = data['Height'].describe()
weight_stats = data['Weight'].describe()

# Calcular cuartiles
height_cuartiles = data['Height'].quantile([0.25, 0.5, 0.75])
weight_cuartiles = data['Weight'].quantile([0.25, 0.5, 0.75])

# Imprimir los resultados
print("Height Estadísticas:")
print(height_stats)
print("\nHeight Cuartiles:")
print(height_cuartiles)
print("\nWeight Estadísticas:")
print(weight_stats)
print("\nWeight Cuartiles:")
print(weight_cuartiles)
```

#### 3. Load country codes

Load the file `../datasets/country_codes.xlsx` which contains a table with country names and their *alpha-2* and *alpha-3* codes.  
Display the table to understand its structure.

```python
file_path = "country_codes.xlsx"
country_codes = pd.read_excel(file_path)

print(country_codes)
```

#### 4. Set index, drop column

Use column `Alpha-2 code` as index. Drop `Alpha-3 code` since it won't be used.

```python
# Establecer "Alpha-2 code" como índice
country_codes.set_index("Alpha-2 code", inplace=True)

# Eliminar columna "Alpha-3 code"
country_codes.drop(columns=["Alpha-3 code"], inplace=True)

print(country_codes)
```

#### 5. Translate country codes in table people into full country names
Merge/join the two tables together in order to add a new column with the country names.

```python
# Código de país como clave de fusión
merged_data = data.merge(country_codes, left_on='Country', right_index=True)

print(merged_data)
```

#### 6. Select the people from Germany

```python
# Seleccionar personas de Alemania
germany_people = merged_data[merged_data['Country'] == 'DE']

print(germany_people)
```

#### 7. Calculate average weight by country

```python
# Calcular el peso promedio por país
average_weight_by_country = merged_data.groupby('Country')['Weight'].mean()

print(average_weight_by_country)
```

#### 8. Calculate minimum and maximum values for height by country
Display minimum and maximum heights per country. Results must appear in a single table.

```python
# Calcular los valores mínimos y máximos de altura por país
height_min_max_by_country = merged_data.groupby('Country')['Height'].agg(['min', 'max'])

print(height_min_max_by_country)
```

#### 9. Calculate averages by country and gender
Let's calculate the average by country and gender at the same time.

#### 9.1. Group by country and gender together
Group by `Country name` and `Gender` in the same `groupby()`, to get aggregated values in two levels.

```python
# Calcular los promedios por país y género
averages_by_country_gender = merged_data.groupby(['Country name', 'Gender']).mean()

print(averages_by_country_gender)
```

#### 9.2. Pivot table
Let's pivot the table to display countries in columns and gender in rows.

```python
# Pivotar tabla
pivot_table = merged_data.pivot_table(index='Gender', columns='Country name', values=['Height', 'Weight'], aggfunc='mean')

print(pivot_table)
```

#### 10. Save results
Save to Excel (or CSV) the pivot table.

```python
# Excel
pivot_table.to_excel("pivot_table.xlsx")

# CSV
pivot_table.to_csv("pivot_table.csv")
```
